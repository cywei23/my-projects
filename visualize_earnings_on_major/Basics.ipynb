## Visualizing Earnings Based On College Majors
In this guided project, we'll explore how using the pandas plotting functionality along with the Jupyter notebook interface allows us to explore data quickly using visualizations. 

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Using visualizations, we can start to explore questions from the dataset like:

- Do students in more popular majors make more money?
     - Using scatter plots
- How many majors are predominantly male? Predominantly female?
    - Using histograms
- Which category of majors have the most students?
    - Using bar plots

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
print(recent_grads.head())
recent_grads.tail()
```

```python
recent_grads.describe()
```

```python
recent_grads.shape[0]
```

```python
recent_grads = recent_grads.dropna(axis=0)
recent_grads.shape[0]
```

One row with missing values was dropped.

### Pandas, Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
```

- Do students in more popular majors make more money?

No. Looks like popular major tends to have lower median salary.

- Do students that majored in subjects that were majority female make more money?

No. More female share tends to have lower median salary.

- Is there any link between the number of full-time employees and median salary?

The more full-time employees, the lower median salary.

### Pandas, Histograms

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(8,40))
for r in range(1, 8):
    ax = fig.add_subplot(8,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40, title=cols[r])
```

```python
fig = plt.figure(figsize=(8,40))
for r in range(1, 8):
    ax = fig.add_subplot(8,1,r)
    ax = recent_grads[cols[r]].hist(bins=25)
    ax.set_title(cols[r])
```

- What percent of majors are predominantly male? Predominantly female?

About 10% of majors are predominantly male (>90% male) and about 10% of majors are predominantly female. 

- What's the most common median salary range?
30 to 40K
